```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv('Appendix_2_Data_for_Figure_2.1.csv')
df.head()
```

```python
df.set_index(['RANK'], inplace=True)
df.drop(df.index[-1], inplace=True)
df.isnull().sum()
```

```python
df.info()
```

```python
df.describe()
```

Data Visualization

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot = True, cmap = 'YlGnBu')
```

```python
top_30_countries = df.sort_values(by='Happiness score', ascending=False).head(30)
top_30_countries
```

```python
for i in range(1, len(top_30_countries.columns)):
    if i == 1 or i >= 5:
        fig = px.bar(top_30_countries, x = top_30_countries['Country'], y = top_30_countries[top_30_countries.columns[i]].values, labels = {'y':top_30_countries.columns[i]}, color = 'Happiness score', width=800)
        fig.show()
        #fig.write_image(f'Happiness_score+{top_30_countries.columns[i]}.png')
```

```python
bottom_30_countries = df.sort_values(by='Happiness score', ascending=True).head(30)
bottom_30_countries = bottom_30_countries.sort_values(by='Happiness score', ascending=False)
bottom_30_countries
```

```python
for i in range(1, len(bottom_30_countries.columns)):
    if i == 1 or i >= 5:
        fig = px.bar(bottom_30_countries, x = bottom_30_countries['Country'], y = bottom_30_countries[bottom_30_countries.columns[i]].values, labels = {'y':bottom_30_countries.columns[i]}, color = 'Happiness score', width=800)
        fig.show()
```

```python
for i in range(1, len(df.columns)):
    if i == 1 or i >= 5:
        plt.figure(figsize=(8, 6))
        sns.displot(df[df.columns[i]], kde = False, bins = 25)
```

```python
df_scatt = df.drop(list(df.columns[i] for i in range(2, 5)), axis=1, inplace=False)
g = sns.PairGrid(df_scatt)
g.map(sns.scatterplot)
```

```python
for i in range(5, len(df.columns)):
    sns.relplot(data = df, x = 'RANK', y = df.columns[i])
```

We can clearly see from the graph that as the rank increases the GDP of that country decreases, but there are some exceptions present.  
Same is the case with Life Expectancy.  
Social support also shows a downward going graph.  
Freedom to make choices is on the upperside for most of the countries.  
Whereas Generosity and Perception of corruption is on the lower side for majority of the Countries.
